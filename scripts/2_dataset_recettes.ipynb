```python
import pandas as pd
'''
Lit un csv qui possède des virgules comme séparateur
'''
data = pd.read_csv("recettes_1.csv", delimiter=",")
data
```

```python
'''
Transforme le csv en DataFrame
'''
from datasets import load_dataset, Dataset

dataset=load_dataset("csv", data_files="recettes_1.csv", split="train")
```

```python
'''
Nettoie le texte pour laisser que les caractères alpha-numériques et les espaces
'''
import spacy
import re

def limpiar_texto(texto):
   
    texto_limpio = re.sub(r'[^a-zA-Z0-9\s]', '', texto)
    return texto_limpio


    
```

```python
'''
Fait des tuples qui possèdent la forme du token et son POS
'''
nlp = spacy.load("es_core_news_sm")
valeurs_textuelles=[]
for fila in dataset:
    for columna, valor in fila.items():
        if isinstance(valor, str):
            texto_limpio=limpiar_texto(valor)
            valeurs_textuelles.append(texto_limpio)

            
resultados_pos = []


for texto in valeurs_textuelles:
    doc = nlp(texto)
    for token in doc:
        resultados_pos.append((token.text, token.pos_))


for resultado in resultados_pos:
    print(resultado)
```
